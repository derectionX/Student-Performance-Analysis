```python
import streamlit as st
import pandas as pd
import joblib

# Load saved model
model = joblib.load("student_model.pkl")

st.title("ðŸŽ“ Student Performance Prediction")

st.write("Fill in the student details to predict exam performance:")

# Input fields for survey data
hours_studied = st.slider("Hours Studied", 0, 12, 5)
attendance = st.slider("Attendance (%)", 0, 100, 80)
parental_involvement = st.slider("Parental Involvement (1=Low, 5=High)", 1, 5, 3)
access_to_resources = st.slider("Access to Resources (1=Poor, 5=Excellent)", 1, 5, 3)
extracurricular = st.slider("Extracurricular Activities (0=None, 5=High)", 0, 5, 2)
sleep_hours = st.slider("Sleep Hours", 0, 12, 7)
previous_scores = st.slider("Previous Scores", 0, 100, 70)
motivation_level = st.slider("Motivation Level (1=Low, 5=High)", 1, 5, 3)
internet_access = st.selectbox("Internet Access", [0, 1])  # 0=No, 1=Yes
tutoring_sessions = st.slider("Tutoring Sessions per week", 0, 10, 2)
family_income = st.slider("Family Income (1=Low, 5=High)", 1, 5, 3)
teacher_quality = st.slider("Teacher Quality (1=Low, 5=High)", 1, 5, 3)
school_type = st.selectbox("School Type", [0, 1])  # 0=Public, 1=Private
peer_influence = st.slider("Peer Influence (1=Negative, 5=Positive)", 1, 5, 3)
physical_activity = st.slider("Physical Activity (hrs/week)", 0, 20, 5)
learning_disabilities = st.selectbox("Learning Disabilities", [0, 1])  # 0=No, 1=Yes
parental_education = st.slider("Parental Education Level (1=Low, 5=High)", 1, 5, 3)
distance_from_home = st.slider("Distance from Home (km)", 0, 50, 5)
gender = st.selectbox("Gender", [0, 1])  # 0=Male, 1=Female

# Put all features into DataFrame
input_data = pd.DataFrame([{
    "Hours_Studied": hours_studied,
    "Attendance": attendance,
    "Parental_Involvement": parental_involvement,
    "Access_to_Resources": access_to_resources,
    "Extracurricular_Activities": extracurricular,
    "Sleep_Hours": sleep_hours,
    "Previous_Scores": previous_scores,
    "Motivation_Level": motivation_level,
    "Internet_Access": internet_access,
    "Tutoring_Sessions": tutoring_sessions,
    "Family_Income": family_income,
    "Teacher_Quality": teacher_quality,
    "School_Type": school_type,
    "Peer_Influence": peer_influence,
    "Physical_Activity": physical_activity,
    "Learning_Disabilities": learning_disabilities,
    "Parental_Education_Level": parental_education,
    "Distance_from_Home": distance_from_home,
    "Gender": gender
}])

# Predict button
if st.button("Predict Performance"):
    prediction = model.predict(input_data)[0]
    st.success(f"ðŸ“Š Predicted Exam Score: {prediction}")
```
